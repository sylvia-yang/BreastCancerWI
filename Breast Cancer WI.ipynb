### Data Cleaning

```python
#check duplicated for cancer_data_means
df_means.duplicated().sum()
```

```python
#check missing values
df_means.info()
```

```python
#rename columns
#remove "_mean" from column names
new_labels = []
for col in df_means.columns:
    if '_mean' in col:
        new_labels.append(col[:-5])  # exclude last 6 characters
    else:
        new_labels.append(col)

# new labels for our columns
new_labels
```

```python
# assign new labels to columns in dataframe
df_means.columns = new_labels

# display first few rows of dataframe to confirm changes
df_means.head()
```

### Drawing Conclusions

```python
df_means['diagnosis']=='M'
```

```python
 #Selecting only malignant records in df_means
df_means_m = df_means[df_means['diagnosis'] == 'M']
df_means_m
```

```python
mask = df_means['diagnosis']=='M'  #step 1: create a mask --> boolean
df_means_m = df_means[mask]        #step 2: make a df with the true values
df_means_m
```

```python
df_means_m['area'].describe()
```

```python
#Selecting only benign records
df_means_b = df_means[df_means['diagnosis']=='B']
df_means_b.head()
```

```python
df_means_b['area'].describe()
```

```python
#Plot the distribution for M and B 
import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots(figsize=(8, 6))
ax.hist(df_means_b['area'], alpha = 0.5, label = 'benign')
ax.hist(df_means_m['area'], alpha=0.5, label='malignant')
ax.set_title('Distributions of Benign and Malignant Tumor Area')
ax.set_xlabel('Area')
ax.set_ylabel('Count')
ax.legend(loc='upper right')
plt.show()
```

## Conclusion: The malignant tutmor generally has larger area size than benign tumors.
